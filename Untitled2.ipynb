```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
groups = 5
Meansofmen= (22, 30, 35, 35, 26)
Meansofwomen = (25, 32, 30, 35, 29)
```

```python
fig, ax = plt.subplots()
index = np.arange(groups)
bar_width = 0.35
```

```python
M = plt.bar(index, Meansofmen, bar_width,color='b',label='Men')
W = plt.bar(index + bar_width, Meansofwomen, bar_width,color='y',label='Women')
plt.xlabel('person')
plt.ylabel('score')
plt.title('barplot')
plt.legend()
plt.show()
```
