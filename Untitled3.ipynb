```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.DataFrame({'country':['United States','Great Britian','China','Russia','Germany'],'gold_medal':[46,27,26,19,17]})
data.to_csv('medal.csv',index=0)
```

```python
df =  pd.read_csv('medal.csv')
```

```python
df
```

```python
country = df["country"]
medal = df["gold_medal"]
```

```python
plt.pie(medal, labels=country)
plt.title("Gold medal achievements of five most successful countries in 2016 Summer Olympics\n")
plt.show()
```
